```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
%matplotlib inline
```

```python
data_prp = pd.read_csv('appdata10.csv')
data_prp
```

## Convert `hour` into Integer

```python
data_prp['hour'] = data_prp['hour'].apply(lambda x: int(x[1:3]))
```

## Convert `first_open` into Timestamp

```python
data_prp['first_open'] = data_prp['first_open'].apply(lambda x: x[:19])
```

```python
data_prp['first_open'] = pd.to_datetime(data_prp['first_open'])
```

## Get year from `first_open`

```python
data_prp['year_fo'] = data_prp['first_open'].dt.year
```

## Get month from `first_open`

```python
data_prp['month_fo'] = data_prp['first_open'].dt.month
```

## Get date from `first_open`

```python
data_prp['day_fo'] = data_prp['first_open'].dt.day
```

```python
data_prp.columns
```

```python
data_prp = data_prp['user first_open year_fo month_fo day_fo dayofweek hour age screen_list numscreens minigame used_premium_feature enrolled enrolled_date liked'.split()]
```

```python
data_prp.head()
```

## Convert `enrolled_date` into Timestamp

```python
data_prp['enrolled_date'] = data_prp['enrolled_date'].apply(lambda x: x if pd.isnull(x) else x[:19]) 
```

```python
data_prp['enrolled_date'] = pd.to_datetime(data_prp['enrolled_date'])
```

```python
data_prp.info()
```

```python
data_prp['enrolled_date'][3] - data_prp['first_open'][3] 
```

## `first_time` to `enrolled_date`

```python
data_prp['fo_to_enrolled'] = data_prp[['first_open', 'enrolled_date']].apply(lambda x: x['enrolled_date']-x['first_open'], axis=1)
```

```python
data_prp
```

```python
data_prp['fo_to_enrolled'].describe()
```

```python
data_prp['fo_to_enrolled'].value_counts()
```

```python
data_prp[data_prp['fo_to_enrolled']=='0 days 00:00:00']
```

## Extract Text in `screen_list`

```python
cvt = CountVectorizer()
cvt
```

```python
extracted_screen = cvt.fit_transform(data_prp['screen_list'])
extracted_screen
```

```python
# get Feature Names
screen_names = cvt.get_feature_names()
print('len : ',len(screen_names))
print(list(screen_names))
```

```python
# get values
extracted_screen.toarray()
```

```python
df_screen_names = pd.DataFrame(extracted_screen.toarray(), columns=cvt.get_feature_names())
df_screen_names
```

## Concat `data` - `df_screen_names`

```python
data = pd.concat([data_prp, df_screen_names],axis=1)
data.head()
```

```python
sizes_data = [len(data[data['enrolled']==0]), len(data[data['enrolled']==1])]
labels = ['not_enrolled', 'enrolled']

plt.pie(sizes_data, labels=labels, autopct='%2.2f%%', startangle=90)
plt.show()
```

```python
print(list(data.columns))
```

```python
features = data.drop(['user', 'first_open', 'screen_list', 'enrolled_date', 'fo_to_enrolled'],axis=1)
```

```python
features
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = features.drop('enrolled',axis=1)
y = features['enrolled']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
sizes = [len(y_train[y_train==0]), len(y_train[y_train==1])]
labels = ['not_enroled', 'enrolled']


plt.title('Train Data')
plt.pie(sizes, labels=labels, autopct='%2.2f%%', startangle=90)

plt.show()
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
```

```python
rfc = RandomForestClassifier()
rfc
```

```python
rfc.fit(X_train,y_train)
```

```python
pred_rfc = rfc.predict(X_test)
proba_rfc = rfc.predict_proba(X_test)
```

```python
print(plot_confusion_matrix(rfc, X_test, y_test))
```

```python
print(confusion_matrix(y_test, pred_rfc))
```

```python
print(classification_report(y_test, pred_rfc))
```

### Gradient Bossting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gradient = GradientBoostingClassifier()
gradient
```

```python
gradient.fit(X_train, y_train)
```

```python
pred_gradient = gradient.predict(X_test)
proba_gradient = gradient.predict_proba(X_test)
```

```python
print(plot_confusion_matrix(gradient, X_test, y_test))
```

```python
print(confusion_matrix(y_test, pred_gradient))
```

```python
print(classification_report(y_test, pred_gradient))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg
```

```python
logreg.fit(X_train,y_train)
```

```python
pred_logreg = logreg.predict(X_test)
proba_logreg = logreg.predict_proba(X_test)
```

```python
print(plot_confusion_matrix(logreg, X_test, y_test))
```

```python
print(confusion_matrix(y_test, pred_logreg))
```

```python
print(classification_report(y_test, pred_logreg))
```
